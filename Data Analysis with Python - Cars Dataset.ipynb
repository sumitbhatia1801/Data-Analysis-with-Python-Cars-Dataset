### Find all the null values in the dataset. If there is a null value in a column, then  fill it with the mean or mode of that column.

```python
# finding count of null values in each column
data.isnull().sum()
```

```python
# some rows have all values as null therefore we are dropping those rows.

data = data[~data["MSRP"].isnull()]
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# filling null values in "Cylinders" column with the mean of that column
data["Cylinders"].fillna(data["Cylinders"].mean() , inplace = True)
```

```python
# All the null values have been removed
data.isnull().sum()
```
